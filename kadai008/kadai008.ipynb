```python
# var が負の整数の場合
var = -3
if var > 0:
  if var%5 == 0 and var%3 == 0:
    print("FizzBuzz")
  elif var%5 == 0:
    print("Buzz")
  elif var%3 == 0:
    print("Fizz")
else:
  print("負の整数")
# var が3の倍数の場合
var = 3
if var > 0:
  if var%5 == 0 and var%3 == 0:
    print("FizzBuzz")
  elif var%5 == 0:
    print("Buzz")
  elif var%3 == 0:
    print("Fizz")
else:
  print("Err")
# var が5の倍数の場合
var = 5
if var > 0:
  if var%5 == 0 and var%3 == 0:
    print("FizzBuzz")
  elif var%5 == 0:
    print("Buzz")
  elif var%3 == 0:
    print("Fizz")
else:
  print("Err")
# var が3と5の公倍数の場合
var = 15
if var > 0:
  if var%5 == 0 and var%3 == 0:
    print("FizzBuzz")
  elif var%5 == 0:
    print("Buzz")
  elif var%3 == 0:
    print("Fizz")
else:
  print("Err")
# var が3と5の倍数以外の場合
var = 8
if var > 0:
  if var%5 == 0 and var%3 == 0:
    print("FizzBuzz")
  elif var%5 == 0:
    print("Buzz")
  elif var%3 == 0:
    print("Fizz")
  else:
    print("Else")
else:
  print("Err")
```
