```python
# クラス
class Human:
  def __init__(self , name , age):
    self.name = name
    self.age  = age

  def check_adult(self):
    print("大人" if self.age >= 20 else "小人")

data = [
      ("太郎",30)
    , ("次郎",25)
    , ("三郎",21)
    , ("四郎",20)
    , ("五郎",19)
]

people = [Human(name,age)for name , age in data]
for i in people:
  i.check_adult()
```
